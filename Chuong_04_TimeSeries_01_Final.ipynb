# Phân rã chuỗi thời gian

## Phân rã theo phương pháp cổ điển

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df1 = df['2016-01-01':'2020-12-31'].copy()
```

```python
df1.head(10)
```

Khi phân rã chuỗi thời gian thì cần phải có dữ liệu đầy đủ của các ngày nên cần phải điền khuyết dữ liệu cho những ngày bị thiếu

```python
df1.Close.count()
```

```python
ts_close = df1.Close.asfreq(pd.offsets.BDay(), method='pad')
```

pd.offsets.BDay() là điền khuyết theo Business day

```python
ts_close.count()
```

Số lượng mẫu đã tăng lên

```python
# Phân rã theo mô hình cộng
close_decompose = seasonal_decompose(ts_close, model='additive')
```

https://www.statsmodels.org/dev/generated/statsmodels.tsa.seasonal.seasonal_decompose.html

Hàm seasonal_decompose phân rã thành 3 thành phần $seasonal$ (bao gồm cả thành phần chu kỳ), $trend$, và $resid$ 

```python
close_decompose.trend[:10]
```

```python
close_decompose.seasonal[:10]
```

```python
close_decompose.resid[:10]
```

```python
close_decompose.observed[:10]
```

```python
import matplotlib as mlt
mlt.rcParams['figure.figsize'] = [16, 8]
fig = close_decompose.plot()
```

# Làm mịn (Smoothing) time series

### Trung bình trượt đơn giản (Simple Moving Average - SMA)
$SMA_t = \frac{\sum_{j=t-w}^{t} x_i}{w}$

```python
sm5 = ts_close.rolling(window=5).mean()
sm20 = ts_close.rolling(window=20).mean()
sm100 = ts_close.rolling(window=100).mean()
```

```python
sm5
```

```python
plt.plot(ts_close, label='Close')
plt.plot(sm5, label='SMA5')
plt.plot(sm20, label='SMA20')
plt.plot(sm100, label='SMA100')
plt.legend()
plt.show()
```

### Trung bình trượt hàm mũ (Exponential moving average - EMA)
\begin{align*}
&ema_t(P,n)\\
&= \beta P_t + \beta (1-\beta)P_{t-1}+\beta(1-\beta )^{2}P_{t-2}+ \cdots\\
&=\beta P_t + (1-\beta) ema_{t-1}(P,n)
\end{align*}

```python
ewm = ts_close.ewm(span=20, adjust=True).mean()
```

# Chuẩn hóa chuỗi thời gian

```python
from sklearn.preprocessing import StandardScaler
```

## Tạo mới một DataFrame chỉ gồm Close đồng thời trích lọc dữ liệu chỉ từ ngày 01/01/2020

```python
close_data = df['2010-01-01':][['Close']].copy()
```

```python
close_data.head()
```

```python
scaler = StandardScaler()
```

```python
scaler.fit_transform(close_data[['Close']])
```

## Tạo cột dữ liệu ngày $t$

```python
close_data[['Close_t']] = scaler.fit_transform(close_data[['Close']])
```

```python
close_data.head()
```

## Tạo cột dữ liệu cho ngày $t-1$

```python

close_data['Close_t1'] = close_data['Close_t'].shift(1)
```

```python
close_data.head()
```

## Tạo cột dữ liệu cho ngày $t-2$ và $t-3$

```python
close_data['Close_t2'] = close_data.Close_t.shift(2)
close_data['Close_t3'] = close_data.Close_t.shift(3)
```

```python
close_data.head()
```

## Xóa những dòng có giá trị NaN

```python

close_data.dropna(inplace=True)
```

```python
close_data.head()
```

# Tạo dữ liệu huấn luyện

```python
from sklearn.model_selection import train_test_split
```

Dự báo ngày $t$ dựa vào ngày $t-1$, $t-2$, $t-3$

```python
features = ['Close_t1', 'Close_t2', 'Close_t3']
target = 'Close_t'
```

```python
X = close_data[features]
y = close_data[target]
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.7, 
                                                    shuffle=True)
```

```python
train_X.head()
```

$shuffle=True$ dữ liệu được lấy xáo trộn thứ tự

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.7, 
                                                    shuffle=False)
```

$shuffle=False$ để giữ lại thứ tự các mẫu dữ liệu (không lấy xáo trộn vì cần theo thứ tự thời gian)

```python
train_X.head()
```

# Dự báo với SVR (Support Vector Regression)
https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html

class sklearn.svm.**SVR**(*, kernel='rbf', degree=3, gamma='scale', coef0=0.0, tol=0.001, C=1.0, epsilon=0.1, shrinking=True, cache_size=200, verbose=False, max_iter=-1)

* **kernel**: {‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’} or callable, default=’rbf’

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
svr_model = SVR(kernel='linear')
svr_model.fit(train_X, train_y)
```

Nên thực nghiệm với các tham số $kernel$ khác nhau

## Kết quả huấn luyện

```python
train_y_pred = svr_model.predict(train_X)
```

```python
mean_absolute_error(train_y, train_y_pred)
```

```python
plt.plot(train_y, label = 'Train Y')
# train_y_pred là một numpy array không có Date index nên cần truyền giá trị trục x là train_y.index
plt.plot(train_y.index, train_y_pred, label = 'Train y pred')
plt.title('Kết quả dự báo trên tập Train')
plt.legend()
plt.show()
```

## Kết quả test

```python
test_y_pred = svr_model.predict(test_X)
mae_test = mean_absolute_error(test_y, test_y_pred)
mae_test
```

```python
plt.plot(test_y, label = 'test_y')
plt.plot(test_y.index, test_y_pred, label = 'test_y_pred')
plt.title('Kết quả dự báo trên tập test')
plt.legend()
plt.show()
```

```python
close_data['Close_pred'] = np.concatenate((train_y_pred, test_y_pred))
```

```python
close_data
```

## Vẽ biểu đồ kết hợp cả train và test

```python
train_size = 0.7
n_train = int(close_data.shape[0]*train_size) # số lượng mẫu huấn luyện
plt.plot(close_data['Close_t'], label='Close', color='blue')
plt.plot(close_data['Close_pred'].iloc[:n_train + 1], label = 'train pred', linestyle='--', color='#ddaa77')
plt.plot(close_data['Close_pred'].iloc[n_train:], label = 'test pred', linestyle='--', color='red')
plt.grid()
plt.legend()
plt.show()
```
